```python
import numpy as np
import pandas as pd
```

```python
traindata = np.genfromtxt("trainboston.csv", delimiter=',')
xtrain = traindata[:,:-1]
ytrain = traindata[:,-1]
```

```python
ytrain.shape
```

```python
from sklearn.ensemble import GradientBoostingRegressor
model = GradientBoostingRegressor()
model.fit(xtrain,ytrain)
```

```python
xtest = np.genfromtxt("testboston.csv", delimiter=',')
predictions = model.predict(xtest)
np.savetxt('bostonpred.csv', predictions, delimiter=',')
```
