```python
import pandas as pd
import numpy as np
```

```python
traindata = np.genfromtxt("ptraining.csv", delimiter=',')
xtrain = traindata[:,:-1]
ytrain = traindata[:, -1]
```

```python
from sklearn.ensemble import GradientBoostingRegressor
model = GradientBoostingRegressor()
model.fit(xtrain, ytrain)
```

```python
xtest = np.genfromtxt('ptesting.csv', delimiter=',')
ypred = model.predict(xtest)
```

```python
np.savetxt('eppredictions.csv', ypred, delimiter= ',')
```
