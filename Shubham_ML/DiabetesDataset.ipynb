```python
import pandas as pd
import numpy as np
```

```python
trainset = np.genfromtxt('trainingdiabetes.csv', delimiter= ",")
xtrain = pd.DataFrame(trainset)
```

```python
ytrain = xtrain[10].copy()
ytrain
```

```python
del xtrain[10]
xtrain
```

```python
from sklearn.linear_model import LinearRegression
alg1 = LinearRegression()
alg1.fit(xtrain, ytrain)
```

```python
testing = np.genfromtxt("testdiabetes.csv", delimiter=",")
xtest = testing.copy()
```

```python
ypred = alg1.predict(xtest)
np.savetxt('predictions.csv', ypred,fmt = '%.5f' ,delimiter= ',')
```
